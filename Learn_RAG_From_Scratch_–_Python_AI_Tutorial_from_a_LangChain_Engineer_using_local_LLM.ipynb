# Cosine_similarity

Cosine similarity is a measure used to determine how similar two vectors are, irrespective of their size.Measure the similarity between documents or text representations.

**1** indicates that the vectors are identical (i.e., they point in the same direction).
**0** indicates orthogonality (i.e., the vectors are at right angles to each other and have no similarity).
**-1** indicates that the vectors are diametrically opposed.

```python
import numpy as np

def cosine_similarity(vec1, vec2):
    dot_product = np.dot(vec1, vec2)
    norm_vec1 = np.linalg.norm(vec1)
    norm_vec2 = np.linalg.norm(vec2)
    return dot_product / (norm_vec1 * norm_vec2)

similarity = cosine_similarity(query_result, document_result)
print("Cosine Similarity:", similarity)
```

```python
import fitz  # PyMuPDF
import re

# Function to load and extract text from the PDF
def load_pdf(file_path):
    # Open the PDF document
    doc = fitz.open(file_path)
    text = ""

    # Iterate through all the pages and extract text
    for page_num in range(doc.page_count):
        page = doc.load_page(page_num)  # Load page
        text += page.get_text() + "\n"  # Extract text from the page and add a newline

    doc.close()
    return text
```

Split documents

```python
# Function to partition the text into chunks of specified size
def partition_text(text, chunk_size=500):
    # Split text by sentences using a regular expression
    sentences = re.split(r'(?<=[.!?]) +', text)

    # Combine sentences into chunks
    chunks = []
    current_chunk = []
    current_length = 0

    for sentence in sentences:
        sentence_length = len(sentence)

        # If adding the next sentence exceeds chunk size, create a new chunk
        if current_length + sentence_length > chunk_size:
            chunks.append(" ".join(current_chunk))
            current_chunk = []
            current_length = 0

        current_chunk.append(sentence)
        current_length += sentence_length

    # Add the last chunk if it's non-empty
    if current_chunk:
        chunks.append(" ".join(current_chunk))

    return chunks
```

```python
# Example usage
file_path = "/content/cymbal-starlight-2024.pdf"
pdf_text = load_pdf(file_path)
text_chunks = partition_text(pdf_text, chunk_size=500)

# Display the number of chunks and a preview of the first chunk
print(f"Total chunks: {len(text_chunks)}")
print(f"First chunk preview: {text_chunks[0]}")
```

# Part 3: Retrieval

**as_retriever** method in Langchain essentially transforms your vector store (in this case, a Chroma vector database) into a retrieval interface.

"K" allows to choose the specific amount of related documents or embedding that you want to pay attention to or take into consideration when you perform the search

```python
retriever = vector_store.as_retriever(search_kwargs={"k": 1})
```

Creat the search or query function

```python
docs = retriever.get_relevant_documents("How should passenger wear seatbelts")
```

```python
from langchain.prompts import ChatPromptTemplate

# Prompt
template = """Answer the question based only on the following context:
{context}

Question: {question}
"""

prompt = ChatPromptTemplate.from_template(template)
prompt
```

Choose LLM

```python
import os
from huggingface_hub import hf_hub_download

from google.colab import userdata
hugging_api = userdata.get('Hugging_rag_api')

# login into hugging face client
from huggingface_hub import login
login(token=hugging_api)

model_id = "openai-community/gpt2"
filesnames = [ "config.json" ,"flax_model.msgpack" ,"merges.txt" ,
              "model.safetensors" ,"pytorch_model.bin" ,"rust_model.ot" ,"tf_model.h5",
              "tokenizer.json" ,"tokenizer_config.json" ,"vocab.json" ,
              "64-fp16.tflite" ,"64-fp16.tflite" ,"64.tflite"
]


# Download LLM Files
for filename in filesnames:
  download_path = hf_hub_download(repo_id=model_id,filename=filename, token = hugging_api)
  print(download_path)


# configure LLM
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForCausalLM.from_pretrained(model_id)

# used model name
text_gen = pipeline("text-generation", model=model,device = -1, tokenizer=tokenizer, max_new_tokens=100)
```

```python
# Create a Langchain-compatible LLM using the Hugging Face pipeline
llm = HuggingFacePipeline(pipeline=text_gen)
```

Chian them together

```python
# Chain
chain = prompt | llm
```

```python
# Run
chain.invoke({"context":docs,"question":"How should passenger wear seatbelts"})
```

RAG chain Automaticalyy

```python
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough

rag_chain = (
    {"context": retriever, "question": RunnablePassthrough()}
    | prompt
    | llm
    | StrOutputParser()
)

rag_chain.invoke("should i wear seatbelts")
```
