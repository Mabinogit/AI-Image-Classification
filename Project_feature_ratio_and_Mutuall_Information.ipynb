# MI between the energy consuption zones

```python
energyconMI = df [['PowerConsumption_Zone1','PowerConsumption_Zone2','PowerConsumption_Zone3']]
energyconMI.head()
```

```python
X = energyconMI.drop('PowerConsumption_Zone2',axis=1)
y = energyconMI['PowerConsumption_Zone2']

```

```python
from sklearn.feature_selection import mutual_info_regression

def mi_scores(X, y, discrete_features):
    mi_scores = mutual_info_regression(X, y, discrete_features=discrete_features)
    mi_scores = pd.Series(mi_scores, name="MI Scores", index=X.columns)
    mi_scores = mi_scores.sort_values(ascending=False)
    return mi_scores
```

```python
mi_scores_result = mi_scores(X, y, discrete_features=[False, False])
print(mi_scores_result)
```

Create Lag Features for Previous Time Steps

To predict energy consumption based on previous patterns, you need to create lag features (for the previous hour, day, week, etc.).

```python
# 30 min lag feature
time_series['energy_lag_30min'] = time_series['PowerConsumption_Zone1'].shift(3)

# Hourly lag feature (predict based on last hour)
time_series['energy_lag_1h'] = time_series['PowerConsumption_Zone1'].shift(6)

# Daily lag feature (predict based on same time yesterday)
time_series['energy_lag_24h'] =time_series['PowerConsumption_Zone1'].shift(144)

# Weekly lag feature (predict based on same time last week)
time_series['energy_lag_1w'] = time_series['PowerConsumption_Zone1'].shift(1008)
```

```python
time_series.head()
```

Rolling Features for Smoothing

You may want to calculate rolling statistics like the average consumption over a window of time (e.g., last 3 hours, last 7 days) to capture trends and smooth short-term fluctuations.

```python
from time import time
# Rolling window for the last 3 hours (hourly average consumption)
time_series['rolling_avg_3h'] = time_series['PowerConsumption_Zone1'].rolling(window=18).mean()

# Rolling window for the last 3 days (daily average consumption)
time_series['rolling_avg_3d'] = time_series['PowerConsumption_Zone1'].rolling(window=432).mean()

# Rolling window for the last 3 weeks (10 min average consumption)
time_series['rolling_avg_3w'] = time_series['PowerConsumption_Zone1'].rolling(window=3024).mean()

# rolling window for the last 30 min of temperautre
time_series['rolling_avg_30min_temp'] = time_series['Temperature'].rolling(window=3).mean()

# rolling window for the last 30 min of general diffuse flow
time_series['rolling_avg_1h_temp'] = time_series['GeneralDiffuseFlows'].rolling(window=6).mean()

# rolling window for the last 30 min of general diffuse flow
time_series['rolling_avg_30min_gen'] = time_series['GeneralDiffuseFlows'].rolling(window=3).mean()

# # rolling window for the last 1 hour of general diffuse flow
time_series['rolling_avg_1h_gen'] = time_series['GeneralDiffuseFlows'].rolling(window=6).mean()
```

Time-Based Features (Day of Week, Hour, etc.)

You can also extract features from the time index to capture seasonality or weekend vs weekday patterns.

```python
# Extracting time-based features
time_series['hour'] = time_series.index.hour
time_series['day_of_week'] = time_series.index.dayofweek
time_series['is_weekend'] = time_series['day_of_week'].isin([5, 6]).astype(int)  # 5=Saturday, 6=Sunday
time_series['month'] = time_series.index.month
```

Create Multi-Horizon Targets (Hour, Day, Week)
Now, youll create your targets (what you want to predict) for different time horizons.

Example for Targets:
Next hour: target_1h

Next day: target_1d

Next week: target_1w

```python
# Predict the next hour's energy consumption
time_series['target_1h'] = time_series['PowerConsumption_Zone1'].shift(-6)

# Predict the next day's energy consumption (24 hours ahead)
time_series['target_1d'] = time_series['PowerConsumption_Zone1'].shift(-114)

# Predict the next week's energy consumption (1 week ahead)
time_series['target_1w'] = time_series['PowerConsumption_Zone1'].shift(-114*7)
```

```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error
from IPython.display import display
from itertools import combinations

class Experiment:
    def __init__(self, features, target, model_class, train_data, test_data, results_df=None, model=None):
        self.features = features
        self.target = target
        self.model = model  # Pre-trained model (optional)
        self.model_class = model_class  # Model class to instantiate if model is None
        self.train_data = train_data
        self.test_data = test_data
        self.results_df = results_df if results_df is not None else pd.DataFrame()
        self.trained_model = None
        self.cv = cv
        self.scoring = scoring

    def train(self):
        """Train the model if not already provided"""
        if self.model is None:
            self.trained_model = self.model_class()  # Instantiate the model
            self.trained_model.fit(self.train_data[self.features], self.train_data[self.target])
        else:
            self.trained_model = self.model  # Use pre-trained model
        return self.trained_model

    def evaluate(self):
        """Evaluate the trained model and update results dataframe"""
        if self.trained_model is None:
            raise ValueError("Model must be trained before evaluation. Call train() first.")

        # Make predictions
        test_copy = self.test_data.copy()
        test_copy['predictions'] = self.trained_model.predict(test_copy[self.features])

        # Filter out rows with missing values
        test_evaluated = test_copy.dropna(subset=[self.target, 'predictions'])

        # Calculate metrics
        mae = mean_absolute_error(test_evaluated[self.target], test_evaluated['predictions'])
        rmse = np.sqrt(mean_squared_error(test_evaluated[self.target], test_evaluated['predictions']))

        # Create new row for results
        new_row_data = {
            'EXP_num': len(self.results_df) + 1,
            'Features': str(self.features),  # Convert list to string for storage
            'Target': self.target,
            'Model': type(self.trained_model).__name__,
            'MAE': mae,
            'RMSE': rmse
        }

        # Update results dataframe
        self.results_df = pd.concat([self.results_df, pd.DataFrame([new_row_data])], ignore_index=True)
        display(self.results_df)

        return {'mae': mae, 'rmse': rmse, 'predictions': test_copy['predictions']}

    def run_experiment(self):
        """Run the complete experiment: train and evaluate"""
        self.train()
        return self.evaluate()


    def run_all_feature_combinations(self, available_features, min_features=1, max_features=None,
                                   display_progress=True, display_final_results=True):
        """
        Run experiments for all possible combinations of features

        Parameters:
        -----------
        available_features : list
            List of all available feature column names
        min_features : int
            Minimum number of features in combination (default: 1)
        max_features : int
            Maximum number of features in combination (default: all features)
        display_progress : bool
            Whether to show progress during experiments
        display_final_results : bool
            Whether to display final results table

        Returns:
        --------
        results_df : DataFrame
            Updated results dataframe with all experiments sorted by performance
        """

        # Set max_features if not specified
        if max_features is None:
            max_features = len(available_features)

        # Generate all combinations
        all_combinations = []
        for r in range(min_features, min(max_features + 1, len(available_features) + 1)):
            all_combinations.extend(list(combinations(available_features, r)))

        if display_progress:
            print(f"Running {len(all_combinations)} feature combinations...")
            print(f"Target: {self.target}")
            print(f"Model: {self.model_class.__name__}")
            print("-" * 50)

        # Store original features to restore later
        original_features = self.features.copy() if isinstance(self.features, list) else [self.features]

        # Run experiment for each combination
        for i, feature_combo in enumerate(all_combinations, 1):
            feature_list = list(feature_combo)

            if display_progress:
                print(f"Experiment {i}/{len(all_combinations)}: {feature_list}")

            try:
                # Update features for this experiment
                self.features = feature_list

                # Reset trained model for new feature set
                self.trained_model = None

                # Run the experiment
                self.run_experiment()

            except Exception as e:
                if display_progress:
                    print(f"   Error with features {feature_list}: {str(e)}")
                continue

        # Restore original features
        self.features = original_features

        if display_progress:
            print("\n" + "="*50)
            print("ALL EXPERIMENTS COMPLETED")
            print("="*50)

        # Sort by performance metric (MAE in ascending order - lower is better)
        self.results_df = self.results_df.sort_values('MAE', ascending=True).reset_index(drop=True)

        if display_final_results:
            print(f"\nTop 10 Feature Combinations (by MAE):")
            display(self.results_df.head(10))

        return self.results_df


    def get_best_features(self, top_n=1, metric='MAE'):
        """
        Get the best performing feature combinations

        Parameters:
        -----------
        top_n : int
            Number of top combinations to return
        metric : str
            Metric to sort by ('MAE' or 'RMSE')

        Returns:
        --------
        DataFrame with top performing feature combinations
        """
        if len(self.results_df) == 0:
            print("No experiments have been run yet.")
            return pd.DataFrame()

        ascending = True  # Lower MAE/RMSE is better
        sorted_results = self.results_df.sort_values(metric, ascending=ascending)

        return sorted_results.head(top_n)




```
