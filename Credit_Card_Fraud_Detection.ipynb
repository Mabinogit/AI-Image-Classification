```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

# Data Analyses and Pre processing

```python
credit = pd.read_csv('/content/creditcard.csv')
```

```python
credit.isnull().sum()
```

```python
credit['Class'].value_counts()
```

```python
legit = credit[credit.Class == 0]
fraud = credit[credit.Class == 1]
```

```python
# STATISTICAL MEASURES OF THE DATA
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare values of both classes/transaction
credit.groupby('Class').mean()
```

**Under-Sampling**
because the distribution of the data is uneven between legit and fraudulent transaction

```python
legit_sample = legit.sample(n=492)
```

```python
# concatanating two data frames
new_credit = pd.concat([legit_sample, fraud], axis=0)

new_credit.describe()
```

```python
new_credit.head()
```

```python
new_credit['Class'].value_counts()
```

```python
x = new_credit.drop(columns='Class', axis=1)
y = new_credit['Class']
```

# Train Model

Train test split

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify=y, random_state=2)
```

Train Model

```python
Model = LogisticRegression()
Model.fit(x_train, y_train)
```

Evaluate Data

```python
accuracy_train = Model.score(x_train, y_train)
print(accuracy_train)
```

```python
accuracy_test = Model.score(x_test, y_test)
print(accuracy_test)
```
