```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

# Data Analyses and Pre-Processsing

Data collection

```python
heart_df = pd.read_csv('/content/heart_disease_data.csv')
```

```python
heart_df.head()
```

```python
heart_df.tail()
```

```python
heart_df.describe()
```

```python
heart_df.isnull().sum()
```

```python
x = heart_df.drop(columns='target', axis=1)
y = heart_df['target']
```

```python
print(y)
```

## Train Model

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify=y, random_state=2)
```

```python
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
accuracy = model.score(x_train, y_train)
print(accuracy)
```

```python
accuracy_test = model.score(x_test, y_test)
print(accuracy_test)
```

# Build System

Input Data

```python

input = (62,0,0,140,268,0,0,160,0,3.6,0,2,2)
input_np = np.asarray(input)
input_reshape = input_np.reshape(1,-1)
```

Output or prediction

```python
prediction = model.predict(input_reshape)
print(prediction)

if (prediction[0]== 0):
  print('The Person does not have a Heart Disease')
else:
  print('The Person has Heart Disease')
```
