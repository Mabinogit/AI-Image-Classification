# Can a Criterion function increase speed of design of a chair by 20% compard to brute force

```python
# computation representation of a chair#

Chair = {
    "seat_height": 45,
    "backrest_angle": 105,
    "seat_depth": 50,
    "leg_length": 30,
    "leg_width": 3
}
```

# Define your criterion function

what good means for you.
for example: A chair is most confortable if seat is between 42 -48 cm backrest is 100 - 110 degres etc

```python


def comfort_score(chair):
  score = 0

  if 42 <= chair["seat_height"] <= 48:
    score += 1
  if 100 <= chair["backrest_angle"] <= 110:
    score += 1


  return score
```

# Define the design space

Decide the range of values for each parameter

```python
seat_height_range = range(20, 90, 1)
backrest_angle_range = range(50, 200,1)
seat_depth_range = range(20, 100,1)
leg_length_range = range(10, 100,1)
leg_width_range = range(1, 10,1)
```

# Implement a brute force search

```python
import itertools
import time

all_combinations = itertools.product(seat_height_range, backrest_angle_range, seat_depth_range, leg_length_range, leg_width_range)

best_score = 0
best_chair = {}

start_time = time.time()

for combination in all_combinations:
  chair = {
    "seat_height": combination[0],
    "backrest_angle": combination[1],
    "seat_depth": combination[2],
    "leg_length": combination[3],
    "leg_width": combination[4]}

  score = comfort_score(chair)

  if score > best_score:
      best_score = score
      best_chair = chair

end_time = time.time()
elapsed_time = end_time - start_time

print(f"Brute-force search completed in {elapsed_time:.2f} seconds")
print(f"Best chair found: {best_chair}")
print(f"Best comfort score: {best_score}")
```

The hypothesis was wrong because criterion formula is an evaluation function not a search function, so it cant soley reduce speed of design
