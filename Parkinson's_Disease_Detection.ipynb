# Data Pre-processing

```python
# Split target from features
target_variable = parkin_df.drop(columns=['name', 'status'], axis=1)
feature_variable = parkin_df['status']
```

Train test split

```python
# train test split
X_train, X_test, Y_train, Y_test = train_test_split(target_variable, feature_variable, test_size=0.2, random_state=2)
```

standardization

```python
# data standardization
scaler = StandardScaler()
scaler.fit(X_train)
x_train = scaler.transform(X_train)
x_test = scaler.transform(X_test)
```

```python
print(x_train)
```

# Train Model

```python
# Use Support Vector Machine
model = svm.SVC(kernel='linear')
model.fit(x_train, Y_train)
```

Test Accuracy of model

```python
# accuracy of model on train data
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction, Y_train)
print(training_data_accuracy)
```

```python
# accuracy of model on test data
x_test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction, Y_test)
print(test_data_accuracy)
```
